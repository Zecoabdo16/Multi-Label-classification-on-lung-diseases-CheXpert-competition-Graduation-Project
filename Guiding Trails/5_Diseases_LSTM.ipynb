```python
import cv2
import numpy as np 
import pandas as pd

import matplotlib.pyplot as plt

import torch
from torch import nn
from torchsummary import summary

import timm
from myGen import MyGen
from libauc.optimizers import PESG
from libauc.losses import AUCM_MultiLabel

from lstmnet import LSTMArch

from sklearn.metrics import roc_auc_score
```

```python
save_weights_path = '/content/drive/MyDrive/Graduation_Project/lstmnet_pretrained_2.pth'
load_weights_path = '/content/drive/MyDrive/Graduation_Project/lstmnet_pretrained.pth'
```

```python
avail_pretrained_models = timm.list_models(pretrained=True)
avail_pretrained_models
```

```python
model = LSTMArch(5)
model = model.cuda()
```

```python
valGen =  MyGen(image_root_path='', 
                 csv_path='valid_5.csv', 
                 image_size=224, 
                 frontal=True,
                 use_clahe=False,
                 n_patch=4,
                 col_to_train=['Cardiomegaly', 'Edema', 'Consolidation', 'Atelectasis', 'Pleural Effusion'])

testloader =  torch.utils.data.DataLoader(valGen, batch_size=32, num_workers=2, shuffle=False)


for data, label in testloader:
    print(data.shape)
    break
```

```python
# dataloader
root = '/content/'

trainGen = MyGen(image_root_path=root, 
                 csv_path='clean_data.csv', 
                 image_size=224, 
                 frontal=True,
                 use_clahe=False,
                 n_patch=4,
                 col_to_train=['Cardiomegaly', 'Edema', 'Consolidation', 'Atelectasis', 'Pleural Effusion'])

valGen =  MyGen(image_root_path=root, 
                 csv_path='valid_5.csv', 
                 image_size=224, 
                 frontal=True,
                 use_clahe=False,
                 n_patch=4,
                 col_to_train=['Cardiomegaly', 'Edema', 'Consolidation', 'Atelectasis', 'Pleural Effusion'])

trainloader =  torch.utils.data.DataLoader(trainGen, batch_size=32, num_workers=2, shuffle=True)
testloader =  torch.utils.data.DataLoader(valGen, batch_size=32, num_workers=2, shuffle=False)

lr = 0.1 
gamma = 1000
imratio = trainGen.imratio_list 
weight_decay = 1e-4
margin = 1.0


# define loss & optimizer
Loss = AUCM_MultiLabel(imratio=imratio, num_classes=5)
optimizer = PESG(model, 
                 a=Loss.a, 
                 b=Loss.b, 
                 alpha=Loss.alpha, 
                 lr=lr, 
                 gamma=gamma, 
                 margin=margin, 
                 weight_decay=weight_decay, device='cuda')

#model.load_state_dict(torch.load(load_weights_path))

# training
#diff_threshold = 0.01
const_threshold = 5

val_auc_mean = 0
best_val_auc = 0
const_cnt = 0
for epoch in range(5):

    for idx, data in enumerate(trainloader):
      train_data, train_labels = data
      train_data, train_labels  = train_data.cuda(), train_labels.cuda()
      y_pred = model(train_data)
      y_pred = torch.sigmoid(y_pred)
      loss = Loss(y_pred, train_labels)

      optimizer.zero_grad()
      loss.backward()
      optimizer.step()
        
      # validation  
      if idx % 400 == 0:
         model.eval()
         with torch.no_grad():    
              test_pred = []
              test_true = [] 
              for data in testloader:
                  test_data, test_labels = data
                  test_data = test_data.cuda()
                  y_pred = model(test_data)
                  test_pred.append(y_pred.cpu().detach().numpy())
                  test_true.append(test_labels.numpy())
            
              test_true = np.concatenate(test_true)
              test_pred = np.concatenate(test_pred)
              val_auc_mean =  roc_auc_score(test_true, test_pred) 
              model.train()

              if best_val_auc < val_auc_mean:
                  const_cnt = 0
                  best_val_auc = val_auc_mean
                  torch.save(model.state_dict(), save_weights_path)
              else:
                  const_cnt += 1            

              #val_diff = val_auc_mean-best_val_auc
              if (const_cnt >= const_threshold):
                  optimizer.update_regularizer(decay_factor=2)
                  model.load_state_dict(torch.load(save_weights_path))
                  const_cnt = 0 

              print ('Epoch=%s, BatchID=%s, Val_AUC=%.4f, Best_Val_AUC=%.4f'%(epoch, idx, val_auc_mean, best_val_auc))
```

```python
root = '/content/'
testGen =  MyGen(image_root_path=root, 
                 csv_path='valid_5.csv', 
                 image_size=224, 
                 frontal=True,
                 use_clahe=False,
                 n_patch=4,
                 col_to_train=['Cardiomegaly', 'Edema', 'Consolidation', 'Atelectasis', 'Pleural Effusion'])

testloader =  torch.utils.data.DataLoader(testGen, batch_size=32, num_workers=2, shuffle=False)
```

```python
model = LSTMArch(5)
model = model.cuda()
model.load_state_dict(torch.load(save_weights_path))
```

```python
model.eval()
with torch.no_grad():    
    test_pred = []
    test_true = [] 
    for data in testloader:
        test_data, test_labels = data
        test_data = test_data.cuda()
        y_pred = model(test_data)
        y_pred = torch.sigmoid(y_pred)
        test_pred.append(y_pred.cpu().detach().numpy())
        test_true.append(test_labels.numpy())
  
    test_true = np.concatenate(test_true)
    test_pred = np.concatenate(test_pred)
```

```python
labels = ['Cardiomegaly', 'Edema', 'Consolidation', 'Atelectasis', 'Pleural Effusion']

results = pd.DataFrame(index=labels)


scores = []
for i in range(5):
    score = roc_auc_score(test_true[:, i], test_pred[:, i])
    scores.append(score)
    
results['AUC'] = scores
```
